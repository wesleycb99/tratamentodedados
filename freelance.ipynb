```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_excel('planilha.xlsx')
```

- Salvando as variaveis de interesse do cliente.

```python
dados_filt = dados[['CNPJ', 'DDD1', 'Telefone1', 'E-mail', 'Razao Social']].copy()
```

```python
dados_filt.head()
```

- Criando uma nova variavel que contenha o DDD e o numero do cliente na mesma coluna.

```python
dados_filt['Telefone'] = dados_filt['DDD1'] + ' ' + dados_filt['Telefone1']
```

```python
dados_filtrado = dados_filt.drop(['DDD1', 'Telefone1'], axis=1)
```

```python
dados_filtrado.head()
```

```python
dados_filtrado.to_excel('dados_filtrados.xlsx', index=False)
```

```python
dados_filtrado.shape
```

- Separando a base de dados em 7, como solicitado pelo cliente.

```python
partes = np.array_split(dados_filtrado, 7)
df1, df2, df3, df4, df5, df6, df7 = partes
```

```python
for i, df_part in enumerate(partes, 1):
    print(f'DataFrame {i} tem {len(df_part)} linhas.')
```

```python
df1.to_excel('df1.xlsx', index=False)
df2.to_excel('df2.xlsx', index=False)
df3.to_excel('df3.xlsx', index=False)
df4.to_excel('df4.xlsx', index=False)
df5.to_excel('df5.xlsx', index=False)
df6.to_excel('df6.xlsx', index=False)
df7.to_excel('df7.xlsx', index=False)
```
